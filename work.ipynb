```python
# Task 1: Loading and Exploring Dataset using pandas in jupiter

import pandas as pd
import os

# Create sample data
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}

# Convert to DataFrame
try:
    df = pd.DataFrame(data)
except FileNotFoundError:
    print("The file was not found. Please check the file path.")
except pd.errors.EmptyDataError:
    print("The file is empty. Please check the file contents.")
except Exception as e:
    print(f"An error occurred: {e}")


# Save as CSV
try:
    if not os.path.exists('sales_data.csv'):
        df.to_csv('sales_data.csv', index=False)
        print("File saved as 'sales_data.csv'.")
    else:
        print("The file 'sales_data.csv' already exists. Not saving.")
except FileNotFoundError:
    print("The file was not found. Please check the file path.")
except pd.errors.EmptyDataError:
    print("The file is empty. Please check the file contents.")
except Exception as e:
    print(f"An error occurred: {e}")

# Load the CSV file
try:
    df = pd.read_csv('sales_data.csv')
except FileNotFoundError:
    print("The file was not found. Please check the file path.")
except pd.errors.EmptyDataError:
    print("The file is empty. Please check the file contents.")
except Exception as e:
    print(f"An error occurred: {e}")

# Display the first 5 rows
print(df.head())

# Show summary of data types and non-null counts
print(df.info())

# Show summary statistics for numeric columns
print(df.describe())

# Count missing values in each column
print(df.isnull().sum())

# Dropping rows with missing values
df_clean = df.dropna()
```

```python
# Task 2: Basic Data Analysis

print(df.describe())
print("Median:\n", df.median(numeric_only=True))
print("Standard Deviation:\n", df.std(numeric_only=True))

grouped = df.groupby('Age').mean(numeric_only=True)
print(grouped)

print("Average age:")
print(df.groupby('Age').mean(numeric_only = True))

print("\nStandard deviation of the ages:")
print(df.groupby('Age').std(numeric_only = True))
```

```python
# Task 3: Data Visualization

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Example dataset
my_data = {'date': pd.date_range(start='2022-01-01', periods=12, freq='M'),
        'sales': [200, 220, 250, 280, 300, 320, 350, 380, 410, 450, 480, 500]}

try:
    df = pd.DataFrame(my_data)
except FileNotFoundError:
    print("The file was not found. Please check the file path.")
except pd.errors.EmptyDataError:
    print("The file is empty. Please check the file contents.")
except Exception as e:
    print(f"An error occurred: {e}")


# Line chart
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['sales'], marker='o', color='b', label='Sales')
plt.title('My Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# Load the iris dataset
try:
    df = sns.load_dataset('iris')
except FileNotFoundError:
    print("The file was not found. Please check the file path.")
except pd.errors.EmptyDataError:
    print("The file is empty. Please check the file contents.")
except Exception as e:
    print(f"An error occurred: {e}")


# Bar chart: average petal length per species
plt.figure(figsize=(10, 6))
sns.barplot(x='species', y='petal_length', data=df, palette='viridis')
plt.title('Average Petal Length per Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Length (cm)')
plt.tight_layout()
plt.show()

# Histogram: distribution of sepal length
plt.figure(figsize=(10, 6))
plt.hist(df['sepal_length'], bins=20, color='g', edgecolor='black')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()

# Scatter plot: sepal length vs. petal length
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal_length', y='petal_length', data=df, hue='species', palette='viridis')
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.show()
```
